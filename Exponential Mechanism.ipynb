```python
import numpy as np
import matplotlib.pyplot as plt

probs = []
eps = 1.0
deltaR = 3.01
rs = np.arange(0,4.0,0.1)
for r in rs:
    if r<=3.01:
        u = r
    else:
        u = 0 
    probs.append(np.exp(eps * u / (2*deltaR)))

%matplotlib inline
plt.plot(rs,probs)
plt.plot([0,3.01],[1,np.exp(3.01/(2*3.01))])
```
